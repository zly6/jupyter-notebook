# 小试牛刀
matplotlib可能是Python 2D绘图领域使用最为广泛的库，它能够让使用者轻松地将数据图形化，并且，提供多样化的输出格式。同时，它几乎能够对图进行所有能够想到的细节修饰


```python
%matplotlib inline
import matplotlib.pyplot as plt #导入绘图包
labels = 'frog', 'hogs', 'dogs', 'logs'  #设定数据标签
sizes = 15, 20, 45, 10 # 设定数据
colors = 'yellowgreen', 'gold', 'lightskyblue', 'lightcoral'
explode = 0, 0.1, 0, 0 # 设定下面饼图pie的模块弹出程度
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',shadow=True, startangle=50)# autopct精度：小数点保留一位
plt.axis('equal')
plt.show()
```

# 函数式绘图
实际上，matplotlib是受MATLAB的启发构建的，MATLAB语言面向过程，利用函数可以轻松绘制图形。
matplotlib有一套完全仿照MATLAB的函数形式的绘图接口，由matplotlib.pyplot模块提供。虽然matplotlib的实现是面向对象，但一般用户只需要使用matplotlib.pyplot就可以完成数据图形绘制，不过，如果涉及到一些特别细节的调整，就不得不去直接调整matplotlib实现的对象。

# 为项目设置matplotlib参数
有两种方法修改参数：
- 使用参数字典rcParams，就像上面对显示中文和负号做的设置一样
- 调用matplotlib.rc()命令，通过传入关键字元祖来修改参数

比较常用的配置项有：
- axis：设置坐标轴边界和表面的颜色、坐标刻度值大小和网格显示
- figure：控制dpi、边界颜色、图形大小、子区的设置
- font：字体集（font family）、字体大小、样式设置
- grid：设置网格颜色和线型
- legend：图例和其中的文本显示
- line：设置线条和标记
- patch：填充2D空间的图形对象，比如多边形和圆，控制线宽，颜色和抗锯齿设置等
- savefig：对保存的图形进行单独设置，例如，设置渲染的文件的背景为白色
- verbose：设置matplotlib在执行期间信息输出，如silent，helpful，debug和debug-anoying
- xticks和yticks：为x,y轴的主刻度和次刻度设置颜色、大小、方向以及标签大小


```python
import numpy as np
# 从 -π-π到+π+π等间隔的256个值
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)
```

```python
plt.plot(X,C)
plt.plot(X,S)
plt.show() 
```

```python
# 创建一个8*6点的图，并设置分辨率为80
plt.figure(figsize=(8, 6), dpi=80)

# 创建一个新的1*1的子图
plt.subplot(1,1,1)

# 绘制曲线，使用蓝色、连续的、宽度为1（像素）的线条
plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")
# 绘制曲线，使用蓝色、连续的、宽度为1（像素）的线条
plt.plot(X, S, color="green", linewidth=1.0, linestyle="-")

# 设置横轴的上下限
plt.xlim(-4.0,4.0)
# 设置横轴坐标点
plt.xticks(np.linspace(-4,4,9,endpoint=True))

# 设置纵轴的上下限
plt.ylim(-1.0,1.0)
# 设置纵轴坐标点
plt.yticks(np.linspace(-1,1,5,endpoint=True))

# 在屏幕上显示
plt.show()
```

```python
# 创建一个8*6点的图，并设置分辨率为80
plt.figure(figsize=(8, 6), dpi=80)

# 创建一个新的1*1的子图
plt.subplot(1,1,1)

# 绘制曲线，使用蓝色、连续的、宽度为1（像素）的线条
plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")
# 绘制曲线，使用蓝色、连续的、宽度为1（像素）的线条
plt.plot(X, S, color="green", linewidth=1.0, linestyle="-")

# 设置横轴的上下限
xmin, xmax = X.min(), X.max()
ymin, ymax = C.min(), C.max()

dx = (xmax - xmin)*0.2
dy = (ymax - ymin)*0.2

plt.xlim(xmin - dx,xmax + dx)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])

# 设置纵轴的上下限
plt.ylim(ymin - dy, ymax + dy)
plt.yticks([-1, 0, +1])

# 在屏幕上显示
plt.show()
```

# 移动脊柱（spines）
坐标轴线和上面的记号连在一起就形成了脊柱（spines），它记录了数值区域的范围，可以放在任何位置
实际中，每幅图有四条脊柱（上下左右），为了将脊柱放在中间，需要将其余的两条（上和右）设为无色，然后调整其他两条

```python
plt.figure(figsize=(8, 6), dpi=80)

xmin, xmax = X.min(), X.max()
ymin, ymax = C.min(), C.max()
dx = (xmax - xmin)*0.2
dy = (ymax - ymin)*0.2
plt.xlim(xmin - dx,xmax + dx)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
plt.ylim(ymin - dy, ymax + dy)
plt.yticks([-1, 0, +1],
          [r'$-1$', r'$0$', r'$+1$'])

#设置轴线
ax = plt.gca()
ax.spines['right'].set_color('none')# 右侧不显示
ax.spines['top'].set_color('none')# 右侧不显示
ax.xaxis.set_ticks_position('bottom')# 设置坐标点在轴下侧
ax.spines['bottom'].set_position(('data',0)) #设置底部轴位置
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-", label='consine')#增加图例
plt.plot(X, S, color="green", linewidth=1.0, linestyle="-", label='sin')
plt.legend(loc='upper left')

plt.show()
```

# 增加注释

```python
plt.figure(figsize=(8, 6), dpi=80)

xmin, xmax = X.min(), X.max()
ymin, ymax = C.min(), C.max()
dx = (xmax - xmin)*0.2
dy = (ymax - ymin)*0.2
plt.xlim(xmin - dx,xmax + dx)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
plt.ylim(ymin - dy, ymax + dy)
plt.yticks([-1, 0, +1],
          [r'$-1$', r'$0$', r'$+1$'])

#设置轴线
ax = plt.gca()
ax.spines['right'].set_color('none')# 右侧不显示
ax.spines['top'].set_color('none')# 右侧不显示
ax.xaxis.set_ticks_position('bottom')# 设置坐标点在轴下侧
ax.spines['bottom'].set_position(('data',0)) #设置底部轴位置
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.plot(X, C, color="blue", linewidth=2.0, linestyle="-", label='consine')#增加图例
plt.plot(X, S, color="green", linewidth=2.0, linestyle="-", label='sin')
plt.legend(loc='upper left')

t = 2*np.pi/3 
plt.plot([t,t],[0,np.cos(t)], color = 'blue', linewidth=2.5, linestyle='--')
plt.scatter([t,],[np.cos(t),], 50, color = 'blue') # 余弦标注点（只有一个点的散点图）
plt.annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$',
            xy=(t, np.cos(t)), xycoords='data',
            xytext=(-90, -50), textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=.2")) # 为余弦注释点写注释

plt.plot([t,t],[0,np.sin(t)], color = 'red', linewidth=2.5, linestyle='--')
plt.scatter([t,],[np.sin(t),], 50, color = 'red') # 正弦标注点（只有一个点的散点图）
plt.annotate(r'$\sin(\frac{2\pi}{3})=\frac{1}{2}$',
            xy=(t, np.sin(t)), xycoords='data',
            xytext=(+10, +30), textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=.2")) # 为余弦注释点写注释

plt.show()
```

# 其他类型的图

## 普通图

```python
n = 256
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y = np.sin(2*X)

plt.axes([0.025, 0.025, 0.95, 0.95]) #画板的大小

plt.plot(X, Y+1, color='blue', alpha=1.00) # alpha 为透明度，越接近1越不透明，越接近0越透明
plt.fill_between(X, 1, Y+1, color='blue', alpha=.25)

plt.plot(X, Y-1, color='blue', alpha=1.00)
plt.fill_between(X, -1, (Y-1), (Y-1) > -1, color='blue', alpha=.25)
plt.fill_between(X, -1, (Y-1), (Y-1) < -1, color='red', alpha=.25)

plt.xlim(-np.pi, np.pi), plt.xticks([])
plt.ylim(-2.5,2.5), plt.yticks([])
plt.savefig('./figures/plot_normal.png', dpi=48) # 存储图像
plt.show()
```

## 散点图 scatter

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
T = np.arctan2(Y,X)

plt.axes([0.025, 0.025, 0.95, 0.95]) #画板的大小

plt.scatter(X,Y, s=75, c=T, alpha=.5)  # s：大小  c: 颜色序列

plt.xlim(-1.5, 1.5), plt.xticks([])
plt.ylim(-1.5,1.5), plt.yticks([])
plt.savefig('./figures/plot_scatter.png', dpi=48) # 存储图像
plt.show()
```
